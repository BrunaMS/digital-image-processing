```python
import cv2 as cv
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img_source = os.path.abspath('images')
img = mpimg.imread(img_source + '/cameraman.png')
```

```python
print(img)
```

```python
fig = plt.imshow(img, cmap = 'gray')
plt.colorbar(fig)

plt.savefig(img_source + '/img_original.jpg')
```

```python
print(img.min())
print(img.max())
```

```python
fig = plt.imshow(img, cmap = 'gray', vmin = 0, vmax = 1)
plt.colorbar(fig)

plt.savefig(img_source + '/img_normalized.jpg');
```
